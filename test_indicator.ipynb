```python
CLOSEprev = df['Close'].shift(periods=1)
HIGHprev = df['High'].shift(periods=1)
LOWprev = df['Low'].shift(periods=1)

# Traditional
df['PP'] = (HIGHprev + LOWprev + CLOSEprev) / 3
df['R1'] = df['PP'] * 2 - LOWprev
df['S1'] = df['PP'] * 2 - HIGHprev
df['R2'] = df['PP'] + (HIGHprev - LOWprev)
df['S2'] = df['PP'] - (HIGHprev - LOWprev)
df['R3'] = df['PP'] * 2 + (HIGHprev - 2 * LOWprev)
df['S3'] = df['PP'] * 2 - (2 * HIGHprev - LOWprev)
df['R4'] = df['PP'] * 3 + (HIGHprev - 3 * LOWprev)
df['S4'] = df['PP'] * 3 - (3 * HIGHprev - LOWprev)
df['R5'] = df['PP'] * 4 + (HIGHprev - 4 * LOWprev)
df['S5'] = df['PP'] * 4 - (4 * HIGHprev - LOWprev)

# Fibonacci
df['PP'] = (HIGHprev + LOWprev + CLOSEprev) / 3
df['R1'] = df['PP'] + 0.382 * (HIGHprev - LOWprev)
df['S1'] = df['PP'] - 0.382 * (HIGHprev - LOWprev)
df['R2'] = df['PP'] + 0.618 * (HIGHprev - LOWprev)
df['S2'] = df['PP'] - 0.618 * (HIGHprev - LOWprev)
df['R3'] = df['PP'] + (HIGHprev - LOWprev)
df['S3'] = df['PP'] - (HIGHprev - LOWprev)

# Classic
df['PP'] = (HIGHprev + LOWprev + CLOSEprev) / 3
df['R1'] = 2 * df['PP'] - LOWprev
df['S1'] = 2 * df['PP'] - HIGHprev
df['R2'] = df['PP'] + (HIGHprev - LOWprev)
df['S2'] = df['PP'] - (HIGHprev - LOWprev)
df['R3'] = df['PP'] + 2 * (HIGHprev - LOWprev)
df['S3'] = df['PP'] - 2 * (HIGHprev - LOWprev)
df['R4'] = df['PP'] + 3 * (HIGHprev - LOWprev)
df['S4'] = df['PP'] - 3 * (HIGHprev - LOWprev)

# Camarilla
df['PP'] = (HIGHprev + LOWprev + CLOSEprev) / 3
df['R1'] = CLOSEprev + 1.1 * (HIGHprev - LOWprev) / 12
df['S1'] = CLOSEprev - 1.1 * (HIGHprev - LOWprev) / 12
df['R2'] = CLOSEprev + 1.1 * (HIGHprev - LOWprev) / 6
df['S2'] = CLOSEprev - 1.1 * (HIGHprev - LOWprev) / 6
df['R3'] = CLOSEprev + 1.1 * (HIGHprev - LOWprev) / 4
df['S3'] = CLOSEprev - 1.1 * (HIGHprev - LOWprev) / 4
df['R4'] = CLOSEprev + 1.1 * (HIGHprev - LOWprev) / 2
df['S4'] = CLOSEprev - 1.1 * (HIGHprev - LOWprev) / 2
```
